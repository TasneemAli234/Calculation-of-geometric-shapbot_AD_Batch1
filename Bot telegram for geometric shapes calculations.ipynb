```python
class AbstractShape(ABC):
    @abstractmethod
    async def area(self, *args):
        pass
    @abstractmethod
    async def perimeter(self, *args):
        pass
    @abstractmethod
    async def volume(self, *args):
        pass

class Circle(AbstractShape):
    def __init__(self, radius):
        self.__radius = radius 
      
    async def area(self):
        return math.pi * (self.__radius ** 2)

    async def perimeter(self):
        return 2 * math.pi * self.__radius
    
    async def volume(self):
        return None 

class Square(AbstractShape):
    def __init__(self, side):
        self.__side = side
    async def area(self):
        return self.__side ** 2

    async def perimeter(self):
        return 4 * self.__side
    
    async def volume(self):
        return None 

class Cube(AbstractShape):
    def __init__(self, side):
        self.__side = side
    
    async def volume(self):
        return self.__side ** 3
    
    async def side_area(self):
        return self.__side ** 2
    
    async def area(self):
        return (6 * (self.__side** 2))
    
    async def perimeter(self):
        return None 
    
class Rectangle(AbstractShape):
    def __init__(self, length, width):
        self.__length = length 
        self.__width = width
    async def area(self):
        return self.__length * self.__width

    async def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    async def volume(self):
        return None 

class Triangle(AbstractShape):
    def __init__(self, side1, side2, side3):
        self.__side1 = side1 
        self.__side2 = side2
        self.__side3 =side3
        
    async def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        return math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
    
    async def perimeter(self):
        return (self.__side1 + self.__side2 + self.__side3)
    
    async def volume(self):
        return None
    
class Trapezoidal (AbstractShape):
    def __init__(self, base1, base2, height, side1, side2):
        self.__base1 = base1 
        self.__base2 = base2
        self.__height = height
        self.__side1 =side1
        self.__side2 = side2
        
    async def area(self):
        return(0.5*(self.__base1 + self.__base2) * self.__height)
    
    async def perimeter(self):
        return(self.__base1 + self.__base2 + self.__side1 + self.__side2)
    
    async def volume(self):
        return None 
    
class Cylinder(Circle):
    def __init__(self, radius, height):
        super().__init__(radius)
        self.__height = height

    async def volume(self):
        return await self.area() * self.__height

    async def area(self):
        base_area = await super().area()
        lateral_area = await self.perimeter() * self.__height  
        return 2 * base_area + lateral_area  
```

Then I created a function to receive user responses, starting with choosing the shape, then receiving the inputs that I asked the user to enter, deal with and analyze, call the appropriate functions for the user's choice of the shape and return the results, which are conditional sentences to examine the user's input and then do the required calculations 

```python
async def request_parameters(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    choice = context.user_data.get('shape')

    if choice is None:
        choice = update.message.text.strip()
        
        if choice in ['1', '2', '3', '4', '5', '6','7']:
            context.user_data['shape'] = choice
            
            if choice == '1':
                await update.message.reply_text("يرجى إدخال نصف القطر:")
            elif choice == '2':
                await update.message.reply_text("يرجى إدخال طول الضلع:")
            elif choice == '3':
                await update.message.reply_text("يرجى إدخال الطول والعرض (مثال: 5 10):")
            elif choice == '4':
                await update.message.reply_text("يرجى إدخال طول الضلع:")
            elif choice == '5':
                await update.message.reply_text(" يرجى إدخال طول الاضلاع 'الضلع الاول والضلع الثاني والضلع الثالث' (مثال2 3 6): ")
            elif choice == '6':
                await update.message.reply_text(" يرجى إدخال طول الاضلاع (طول ضلع القاعدة1 ووطول القاعدة2والارتفاع والضلع الثالث وطول الصلع الرابع): (مثال2 3 6 7 8): ")
            elif choice == '7':
                await update.message.reply_text("يرجى ادخال نصف القطر والارتفاع (مثال 20 9):")
        else:
            await update.message.reply_text("خيار غير صالح. يرجى إعادة المحاولة.")
    else:
        shape = choice
        try:
            if shape == '1':  # circle
                radius = float(update.message.text)
                if radius > 0:
                    circle = Circle(radius)
                    area = await circle.area()
                    perimeter = await circle.perimeter()
                    await update.message.reply_text(f"قوانين حساب مساحة ومحيط الدائرة:\n المساحة = π × (نصف القطر)^2\n المحيط = 2 × π × نصف القطر")
                    await update.message.reply_text(f"مساحة الدائرة: {area:.2f}\nمحيط الدائرة: {perimeter:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text("يرجى إدخال قيمة موجبة لنصف القطر:")
                    return 
            
            elif shape == '2':  # square
                side = float(update.message.text)
                if side > 0:
                    square = Square(side)
                    area = await square.area()
                    perimeter = await square.perimeter()
                    await update.message.reply_text(f"قوانين مساحة ومحيط المربع:\nالمساحة = طول الضلع ^ 2\n المحيط = 4 × طول الضلع")
                    await update.message.reply_text(f"مساحة المربع: {area:.2f}\nمحيط المربع: {perimeter:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text("يرجى إدخال قيمة موجبة لطول الضلع:")
                    return
            
            elif shape == '3':  # rectangle
                length, width = map(float, update.message.text.split())
                if length > 0 and width > 0:
                    rectangle = Rectangle(length,width)
                    area = await rectangle.area()
                    perimeter = await rectangle.perimeter()
                    await update.message.reply_text(f"قوانين مساحة ومحيط المستطيل:\nالمساحة = الطول × العرض\n المحيط = 2 × (الطول + العرض)")
                    await update.message.reply_text(f"مساحة المستطيل: {area:.2f}\nمحيط المستطيل: {perimeter:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text("يرجى إدخال قيم موجبة للطول والعرض:")
                    return
            
            elif shape == '4':  # cube
                side = float(update.message.text)
                if side > 0:
                    cube = Cube(side)
                    volume = await cube.volume()
                    Side_area = await cube.side_area()
                    Area = await cube.area()
                    await update.message.reply_text(f"قوانين مساحة ومحيط والمساحة الجانبية للمكعب:\nالحجم = طول الضلع ^ 3\nالمساحة الكلية للمكعب = 6× الضلع^2\n المساحة الجانبية = طول الضلع ^2")
                    await update.message.reply_text(f"حجم المكعب: {volume:.2f}")
                    await update.message.reply_text(f"مساحة المكعب الجانبية:{Side_area:.2f}") 
                    await update.message.reply_text(f"مساحة المكعب: {Area:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text("يرجى إدخال قيمة موجبة لطول الضلع:")
                    return
            elif shape == '5':  # tringle
                side1, side2, side3 = map(float, update.message.text.split())
                if side1> 0 and side2> 0 and side3 > 0:
                    tringle = Triangle(side1, side2, side3)
                    area = await tringle.area()
                    perimeter = await tringle.perimeter()
                    await update.message.reply_text(f"قوانين مساحة ومحيط المثلث:\nالمساحة = (1/2) × القاعدة × الارتفاع\n المحيط = طول الضلع الأول + طول الضلع الثاني + طول الضلع الثالث")
                    await update.message.reply_text(f"مساحة المثلث: {area:.2f}\nمحيط المثلث: {perimeter:.2f}")
                    del context.user_data['shape']
            
            elif shape == '6':  # trapezoidal
                base1, base2, height, side1, side2 = map(float, update.message.text.split())
                if  base1 > 0 and base2 > 0 and height> 0 and side1 > 0 and side2 > 0:
                    trapezoidal = Trapezoidal(base1, base2, height, side1, side2)
                    area = await trapezoidal.area()
                    perimeter = await trapezoidal.perimeter()
                    await update.message.reply_text(f"قوانين مساحة ومحيط شبة المنحرف:\nالمساحة = (1/2) × (طول القاعدة الأولى + طول القاعدة  الثانية) × الارتفاع\n المحيط = طول القاعدة الأولى+طول القاعدة الثانية+طول الضلع الأول+طول الضلع الثاني")
                    await update.message.reply_text(f"مساحة شبة المنحرف :{area:.2f}\nمحيط المنحرف: {perimeter:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text(" يرجى إدخال قيم موجبة للارتفاع والقاعدة1و القاعدة2و الضلع الثالث والضلع الرابع:")
                    return
            elif shape == '7':  # cylinder
                radius, height = map(float, update.message.text.split())
                if radius > 0 and height > 0:
                    cylinder = Cylinder(radius,height)
                    area = await cylinder.area()
                    volume = await cylinder.volume()
                    await update.message.reply_text(f"قوانين حساب المساحة والحجم للاسطوانة:\n   المساحة = π × 2 × نصف القطر × (نصف القطر + الارتفاع)\n الحجم = π × (نصف القطر ^ 2) × الارتفاع")
                    await update.message.reply_text(f"مساحة الاسطوانة: {area:.2f}\nحجم الاسطوانة: {volume:.2f}")
                    del context.user_data['shape']
                else:
                    await update.message.reply_text("يرجى إدخال قيم موجبةلنصف القطر والارتفاع :")
                    return   
        except ValueError:
            await update.message.reply_text("يرجى إدخال قيمة صحيحة.")
```

Then I created the main function that builds the interaction with the bot by adopting the bot that you created using (BOTFather) and obeys us (token) that creates the bot The program calls the (start) function and then calls the (request_parameters) function, and then we write the beginning of the interaction with the bot The user enters (start)  
     and the program starts  

```python
async def main():
    app = ApplicationBuilder().token("token that you gat it from BOtFather").build()

    app.add_handler(CommandHandler("start", start))
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, request_parameters))
    

    await app.run_polling()

if __name__ == '__main__':
    import asyncio
    asyncio.run(main())
```
